### As quiet evident from the dataset, MJD represents the Modified Julian Date which is the number of days since midnight on November 17, 1858.

### RA is the right ascension and Dec is the declination with their corresponding errors in the adjacent fields.

### The RA and Dec fields are in form of degree:min:sec. In order to plot them efficiently we would convert them to Decimal Degree.

```
# function to convert the angles in deg:min:sec to corresponding Decimal Degrees
def DecimalDegree(angle):
  deg,min,sec=angle.split(':')
  deg=float(deg)
  min=float(min)
  sec=float(sec)
  return deg+min/60+sec/3600
```

```
# Converts the RA and Dec fields to Decimal Degree
data["RA"]=data["RA"].apply(DecimalDegree)
data["Dec"]=data["Dec"].apply(DecimalDegree)
#Warning:Running this cell more than once will throws error since this function also involves typecasting.
```

```
data
```

# Plotting the errors

#### This doesn't has much significance but just an analysis to visualise the error distribution incase it turns helpful later in analysing some aspects.

```
plt.subplot(2,1, 1)
plt.plot(data.RA, data.RAs_er, 'ro')
plt.title('Err in RA with increasing RA')  
plt.xlabel('RA')  
plt.ylabel('RA err')  

plt.subplot(2,1, 2)
plt.plot(data.Dec, data.Decs_er, 'ro')
plt.title('Err in Dec with increasing Dec')  
plt.xlabel('Dec')  
plt.ylabel('Dec err')

plt.tight_layout()

plt.show()
```

# Plotting Declination with RA

```
data.plot(x='RA', y='Dec', style='o')  
plt.title('Dec with RA')  
plt.xlabel('RA')  
plt.ylabel('Dec')  
plt.show()
```

# Plotting Right Ascension and Declination with time(in MJD)

```
data.plot(x='MJD', y='RA', style='o--')  
plt.title('RA with MJD')  
plt.xlabel('MJD')  
plt.ylabel('RA')  
plt.show()
```

```
data.plot(x='MJD', y='Dec', style='o--')  
plt.title('Dec with MJD')  
plt.xlabel('MJD')  
plt.ylabel('Dec')  
plt.show()
```
